#### Testing pairwise

```python
print('If p value is less than 0.05, Wilcoxon rank sum test shows there is significant difference in means')
sign = []
norm_counts_tot = []
AUC_tot = []
# n_test = 41 #Total
n_test = 34 #No Dieter, no all
for k in range(n_test):
    count = 1

    #Smaller tests
    data = [(data_copy[k]), (data_copy[k+n_test])]
    names = [names_copy[k], names_copy[k+n_test]]
    names_tot.append(names)
    print('Analyzing: ', names)
    if names[0] == 'CA_d_all' or  names[0] == 'CA_b_all' or  names[0] == 'CA_e_all' or names[0] =='CA_LBFGS_all' or names[0] =='CA_VN_all' or names[0] =='CA_LBFGS_D':
        print(names, len(data[0]),len(data[1]))
        #Fixing different number of samples
        data = [data[0],data[1][:len(data[0])]]

    #Next, check ANOVA (or Krushal Wallis if non parametric): https://www.marsja.se/four-ways-to-conduct-one-way-anovas-using-python/
    #st, p_a = scipy.stats.f_oneway(*data)
    # print('If p value is less than 0.05, ANOVA shows there is significant difference. P-value: ', p_a, '. Stat: ', st)

    #If they were parametric data we could do t test
    #st,p_w = scipy.stats.ttest_ind(data[0],data[1])
    
    #Note, the ones that are parametric, have non significant differences, The applicable one in our case is Krushal Wallis
    #h, p_k = scipy.stats.kruskal(*data)
    #print('P-value k: ', p_k, '. Stat: ', h)
    #Dunn's test
    #results = sp.posthoc_dunn(data, p_adjust = 'bonferroni')
    #print('Significant difference dunn''s: ', results.values)
    
    #Conover's test
    #results = sp.posthoc_conover(data, p_adjust = 'bonferroni')
    #print('Significant difference conover''s: ', results.values)
    
    #Doing the actual test for only 2 groups: Wilcoxon
    #This version is for paired data (Wilcoxon signed rank), not our case
    #st_w, p_w = scipy.stats.wilcoxon(data[0],data[1])
    #print('P-value wilcoxon: ', p_w, '. Stat: ', st_w)
    #This version is the Wilcoxon rank sum aka mann whitney u test
    st_m,p_m = scipy.stats.mannwhitneyu(data[0],data[1])
    print('P-value mannwhiteney: ', p_m, '. Stat: ', st_m)
    sp, p_sp = scipy.stats.spearmanr(data[0],data[1])
    print('Spearman correlation: ', sp, p_sp)
    #The p-value for a hypothesis test whose null hypotheisis is that two sets of data are uncorrelated.
    
    p_tot.append([p_m,sp])
    
    #Computing F1 score if it was a classification problem: Challengue bc we would have to chose the threshold discretization
    f1_FA = []
    cont_FA = []
    f1_CA = []
    cont_CA = []
    for i in range(len(data[1])):
        # First looking at the amount of false negatives that would appear
        cont_1 = data[1][i]
        fn = np.sum(data[0] <= cont_1)
        tp = np.sum(data[0] > cont_1)
        fp = np.sum(data[1] > cont_1)
        # print('Should be total elements: ', (tp + fn))
        # Precision: True pos / (true pos + false pos)
        prec = tp / (tp + fp)
        # Recall: True pos/ (true pos + false neg)
        rec = tp / (tp + fn)
        # F1 score: 2*(precision*recall)/(precision + recall)
        f1_fa = 2 * (prec * rec) / (prec + rec)
        f1_FA.append(f1_fa)
        cont_FA.append(cont_1)
        # Next looking at the amount of false postives that would appear
        cont_2 = data[0][i]
        fp = np.sum(data[1] >= cont_2)
        fn = np.sum(data[0] <= cont_2)
        tp = np.sum(data[0] > cont_2)
        # print('Should be total elements: ', (tp + fn))
        # Precision: True pos / (true pos + false pos)
        prec = tp / (tp + fp)
        # Recall: True pos/ (true pos + false neg)
        rec = tp / (tp + fn)
        # F1 score: 2*(precision*recall)/(precision + recall)
        f1_ca = 2 * (prec * rec) / (prec + rec)
        f1_CA.append(f1_ca)
        cont_CA.append(cont_2)
    F1 = max([max(f1_CA), max(f1_FA)])
    id_1 = np.nanargmax([max(f1_CA), max(f1_FA)])
    if np.sum(np.shape(id_1))> 1 :
        id_1 = id_1[0]
    id_2 = [np.nanargmax(f1_CA), np.nanargmax(f1_FA)]
   # if np.sum(np.shape(id_2))> 1 :
   #     id_2 = id_2[0]
    cont = [cont_CA, cont_FA]
    cont_max = cont[id_1][id_2[id_1]]
    #Best F1, 1; worst F1 0
    print('F1 score: ', F1, ' at: ', cont_max)
    
    #Computing error metric
    count = 0
    for i in range(len(data[0])):
        for j in range(len(data[1])):
            if data[0][i]<data[1][j]:
                count+=1
    print('Error metric: ' ,count, ' Normalized :', count/(len(data[0])*len(data[1])), ' total data: ', len(data[0])*len(data[1]))
    
    #Computing ROC curve of one pair
    kr = np.concatenate([data[0], data[1]])
    label = np.concatenate([np.ones((len(data[0]))), np.zeros((len(data[1])))])
    fpr, tpr, thr = roc_curve(label, kr)
    roc_auc = auc(fpr, tpr)
    a = []
    maxs = np.argmax(tpr + 1 - fpr)
    if np.sum(np.shape(maxs)) >= 1:
        for i in range(np.sum(np.shape(maxs)) >= 1):
            a.append(tpr[i])
        id = np.argmax(a)
        maxs = maxs[id]
    #Computing F1 score based on ROC criterion
    cont = maxs
    fn = np.sum(data[0] <= cont)
    tp = np.sum(data[0] > cont)
    fp = np.sum(data[1] > cont)
    # print('Should be total elements: ', (tp + fn))
    # Precision: True pos / (true pos + false pos)
    prec = tp / (tp + fp)
    # Recall: True pos/ (true pos + false neg)
    rec = tp / (tp + fn)
    # F1 score: 2*(precision*recall)/(precision + recall)
    F1_ROC = 2 * (prec * rec) / (prec + rec)
    print('F1 score ROC: ', F1_ROC)

    
#     if p_m < 0.05 or F1 > or names[0] == 'CA_LBFGS' or names[0] == 'CA_LBFGS_rand' or names[0] == 'CA_VN_all'  or names[0] == 'CA_VN' or names[0] == 'CA_VN_rand':     
            
    fig2, ax2 = plt.subplots(1,2,figsize = (9,3))
    ax2[0].plot(fpr, tpr)
    ax2[0].plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
    ax2[0].set_xlim([-0.05, 1.0])
    ax2[0].set_ylim([0.0, 1.05])
    ax2[0].set_xlabel('1 -Specificity (FPR)',size = 'small')
    ax2[0].set_ylabel('Sensitivity (TPR)',size = 'small')
    ax2[0].grid('on')
    ax2[0].set_title('ROC curve, AUC: ' + str(np.round(roc_auc, decimals=3)) + '\n Best criterion ' + str(np.round(thr[maxs],decimals = 3)) +'[m/s] : Sens ' + str(np.round(tpr[maxs],decimals = 3)) + ' Spec ' + str(np.round(1- fpr[maxs],decimals = 3)) ,size='small')

    #Plotting the data
    ax2[1].boxplot(data)
    ax2[1].set_xticklabels(names, size = 'small')
    ax2[1].violinplot(data)
    for i in range(len(data)):
        ax2[1].text(i + 1, 4, np.around(statistics.median(data[i]), decimals=4),
#                         rotation=45,
                    ha='left', size = 'x-small')
#         ax2[1].tick_params(labelrotation=45)
    ax2[1].set_title('SoS distribution' '\n p value: ' + str(np.around(p_m, decimals=4)) + ' F1 score: ' + str(np.around(F1, decimals=2)) + ' Error: ' + str(np.round(count/(len(data[0])*len(data[1])),decimals=2)),size='small')
    ax2[1].set_ylabel('SoS [m/s]')
    #plt.ylim(0,9)
#     pk = 0
#     while os.path.isfile('/scratch_net/biwidl307/sonia/kk/figures/dilated_case_study_all' + str(k) + str(pk) + '.pdf'):
#         pk += 1
#     plt.savefig('/scratch_net/biwidl307/sonia/kk/figures/dilated_case_study_all' + str(k) + str(pk) + '.pdf')
    plt.show()
    sign.append([np.round(p_m,decimals = 5), np.round(roc_auc, decimals = 4), np.round(F1, decimals = 4), np.round(F1_ROC, decimals = 4),  count,np.round(count/(len(data[0])*len(data[1])),decimals=4),names])
    norm_counts_tot.append(count/(len(data[0])*len(data[1])))
    AUC_tot.append(roc_auc)    
```

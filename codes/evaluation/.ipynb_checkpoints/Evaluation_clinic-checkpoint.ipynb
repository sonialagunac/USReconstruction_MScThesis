```python
def sample_plots(experiment,samples, mat, uncert, vn, MS):
    # Plotting different examples of different frames
    ind = []
    rmse = []
    rmse_sub = []
    ind_c = []
    rmse_c = []
    rmse_sub_c = []
    #Ranking based on deltaSoS
    for n in range(uncert['samples'].shape[0]):
        for m in range(uncert['samples'].shape[1]):
            if n == 0:
                rmse_sub.append([])
                rmse_sub_c.append([])
            pre = np.transpose(uncert['samples'][n, m, ...])
            sig_in = np.mean(pre[np.where((uncert['mask'][...,m]) == 1)])
            sig_out = np.mean(pre[np.where((uncert['mask_out'][...,m]) == 1)])
            dev_out = np.std(pre[np.where((uncert['mask_out'][...,m]) == 1)])
            dev_in = np.std(pre[np.where((uncert['mask_out'][...,m]) == 1)])
            #CNR = np.abs(sig_in - sig_out) / np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2))
            CNR_c = np.abs(sig_in - sig_out) / sig_out
            CNR = np.abs((sig_in - sig_out))
            rmse_sub[m].append(CNR)
            rmse_sub_c[m].append(CNR_c)
    for m in range(uncert['samples'].shape[1]):
        ind.append(sorted(range(len(rmse_sub[m])), key=lambda k: rmse_sub[m][k]))
        ind_c.append(sorted(range(len(rmse_sub_c[m])), key=lambda k: rmse_sub_c[m][k]))
 
    #In case there is no mask available
    #rmse_sub = []
    #ind = []
    #for n in range(uncert['samples'].shape[1]):
    #    rmse_sub.append(np.ndarray.tolist(np.arange(0,uncert['samples'].shape[0],1)))
    #    ind.append(np.ndarray.tolist(np.arange(0, uncert['samples'].shape[0], 1)))
    if samples > 0:
        for num in np.linspace(0, uncert['mean'].shape[0] - 1, samples):
            num = int(num)
            #vmin_c = np.amin([uncert['mean'][num, ...], vn['recon'][num, ...]])
            #vmax_c = np.amax([uncert['mean'][num, ...], vn['recon'][num, ...]])
            vmin_c = 1460
            vmax_c = 1540
            vmin_st = 2
            vmax_st = 20
            f, axarr = plt.subplots(2, 4, figsize=(
                10, 5 * uncert['mean'][num, ...].shape[0] / uncert['mean'][num, ...].shape[1]))
            #Reconstruction mean
            a0 = axarr[0, 0].imshow(np.transpose(uncert['mean'][num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            axarr[0, 0].set_title('Mean [m/s]', fontsize=13)
            axarr[0, 0].set_xticks([])
            axarr[0, 0].set_yticks([])
            p0 = plt.colorbar(a0, ax=axarr[0, 0], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p0.ax.tick_params(labelsize=10)
            #Uncertainty
            a1 = axarr[0, 1].imshow(np.transpose(uncert['stdev'][num, ...]), cmap='hot_r', vmin = vmin_st, vmax = vmax_st)
            axarr[0, 1].set_title('St. dev [m/s]',
                fontsize=13)
            axarr[0, 1].set_xticks([])
            axarr[0, 1].set_yticks([])
            p1 = plt.colorbar(a1, ax=axarr[0, 1], fraction=0.046 * (uncert['mean'][num, ...].shape[1] /uncert['mean'][num, ...].shape[0]))
            p1.ax.tick_params(labelsize=10)
            #Bmode image
            b_mode_im_pre = np.abs(hilbert(np.mean(uncert['b_mode'][..., num], 2)))
            b_mode_im = 20 * np.log10(b_mode_im_pre / np.max(np.abs(b_mode_im_pre)))
            if MS:
                a2 = axarr[1, 0].imshow(b_mode_im, aspect=3.75/15, cmap='Greys_r')
            else:
                a2 = axarr[1, 0].imshow(b_mode_im, aspect=3.75/30, cmap='Greys_r')
            for im in axarr[1,0].get_images():
                im.set_clim(-70, 0)
            axarr[1, 0].set_title('B mode image [dB]')
            p2 = plt.colorbar(a2, ax=axarr[1, 0], fraction=0.046 * (
                uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p2.ax.tick_params(labelsize=10)
            axarr[1, 0].axis('off')
            #Max CNR sample
            a3 = axarr[1, 1].imshow(np.transpose(uncert['samples'][ind[num][len(ind[num])-1],num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            p3 = plt.colorbar(a3, ax=axarr[1, 1], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            axarr[1, 1].set_title('Max CNR ' + str(np.around(rmse_sub[num][ind[num][len(ind[num]) - 1]], decimals=2)) + 'exp ' + str(ind[num][len(ind[num])-1]),
                                      fontsize=13)
            axarr[1, 1].set_xticks([])
            axarr[1, 1].set_yticks([])
            p3.ax.tick_params(labelsize=10)
            #Min CNR sample
            a4 = axarr[0, 2].imshow(np.transpose(uncert['samples'][ind[num][0], num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            p4 = plt.colorbar(a4, ax=axarr[0, 2], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            axarr[0, 2].set_title('Min CNR ' + str(np.around(rmse_sub[num][ind[num][0]], decimals=2))+ ' exp ' + str(ind[num][0]), fontsize=13)
            axarr[0, 2].set_xticks([])
            axarr[0, 2].set_yticks([])
            p4.ax.tick_params(labelsize=10)
            #0.25 CNR sample
            a5 = axarr[0, 3].imshow(np.transpose(uncert['samples'][ind[num][int(0.25*(len(ind[num])-1))],num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            p5 = plt.colorbar(a5, ax=axarr[0, 3], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            axarr[0, 3].set_title(
                    '0.25 CNR ' + str(np.around(rmse_sub[num][ind[num][int(0.25 * (len(ind[num]) - 1))]], decimals=2))+ ' exp ' + str(ind[num][int(0.25*(len(ind[num])-1))]),
                    fontsize=13)
            axarr[0, 3].set_xticks([])
            axarr[0, 3].set_yticks([])
            p5.ax.tick_params(labelsize=10)
            #0.5 CNR sample
            a6 = axarr[1, 3].imshow(np.transpose(uncert['samples'][ind[num][int(0.5*(len(ind[num])-1))],num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            p6 = plt.colorbar(a6, ax=axarr[1, 3], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            axarr[1, 3].set_title('0.5 CNR ' + str(np.around(rmse_sub[num][ind[num][int(0.5*(len(ind[num])-1))]],decimals = 2))+ ' exp ' + str(ind[num][int(0.5*(len(ind[num])-1))]), fontsize=13)

            axarr[1, 3].set_xticks([])
            axarr[1, 3].set_yticks([])
            p6.ax.tick_params(labelsize=10)
            #0.75 CNR sample
            a7 = axarr[1, 2].imshow(np.transpose(uncert['samples'][ind[num][int(0.75*(len(ind[num])-1))],num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            p7 = plt.colorbar(a7, ax=axarr[1, 2], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            axarr[1, 2].set_title('0.75 CNR ' + str(np.around(rmse_sub[num][ind[num][int(0.75 * (len(ind[num]) - 1))]], decimals=2))+ ' exp ' + str(ind[num][int(0.75*(len(ind[num])-1))]),
                    fontsize=13)
            axarr[1, 2].set_xticks([])
            axarr[1, 2].set_yticks([])
            p7.ax.tick_params(labelsize=10)
            #Extras, mode
            # mode = scipy.stats.mode(uncert['samples'][:, num, :], axis=0)
            # a5 = axarr[0, 3].imshow(np.transpose(mode.mode[0,...]), cmap='turbo', vmin=vmin_c,
            #                         vmax=vmax_c)
            # p5 = plt.colorbar(a5, ax=axarr[0, 3], fraction=0.046 * (
            #         uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            # axarr[0, 3].set_title('Mode [m/s]')
            # axarr[0, 3].set_xticks([])
            # axarr[0, 3].set_yticks([])
            # p5.ax.tick_params(labelsize=10)
            f.subplots_adjust(wspace=0.3)
            f.suptitle(experiment[0])
            plt.show()
    return rmse_sub, ind, rmse_sub_c, ind_c
```

```python
def compute_corr(uncert, vn, lbfgs_recon, data_names,stdev_loc_tot,stdev_tot,MS, CNR_tot,nCNR_tot,CNR_lbfgs,nCNR_lbfgs,CNR_VN,nCNR_VN, plot = False, samples = 3):
    uncert['mask_dev'] = []
    for num in np.linspace(0,uncert['mean'].shape[0] - 1,samples):
        num = int(num)
        #Computing L*Meas
        b = np.ones(np.shape(bf_im['measmnts'][:, num]))
        b[np.isnan(bf_im['measmnts'][:, num])] = 0
        #Computing L*CorrCoeff
        # b = orig_data['CorrCoeff'][:, num].copy()
        # b[np.isnan(orig_data['CorrCoeff'][:, num])] = 0
        a = np.matmul(np.transpose(np.abs(bf_im['L'])), b)
        backp = (a.reshape(84, 64, order='F').copy())
        backp = (backp - np.amin(backp[7:, :])) / (np.amax(backp[7:, :]) - np.amin(backp[7:, :]))
        if short:
            backp = backp[:67,...]
        kk = np.transpose(uncert['stdev'][num,:, 7:]) * (1 - 0.5*backp[7:, :])
        mask_dev = np.mean(kk[uncert['mask'][7:67,:,num] == 1])
        uncert['mask_dev'].append(mask_dev)

        if plot:
            #Computing Bmode intensities
            b_mode_im_pre = np.abs(hilbert(np.mean(uncert['b_mode'][..., num], 2)))
            b_mode_inter = b_mode_im_pre / np.max(np.abs(b_mode_im_pre))
            b_mode_im = 20 * np.log10(b_mode_im_pre / np.max(np.abs(b_mode_im_pre)))

            vmin_c = 1460
            vmax_c = 1540
            vmin_st = 2
            vmax_st = 20
            # vmin_c = np.amin([uncert['mean'][num, ...], vn['recon'][num, ...], np.transpose(lbfgs_recon[num,...])])
            # vmax_c = np.amax([uncert['mean'][num, ...], vn['recon'][num, ...], np.transpose(lbfgs_recon[num,...])])

            f, axarr = plt.subplots(2, 3)
            #Mean recon
            a0 = axarr[0, 0].imshow(np.transpose(uncert['mean'][num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            axarr[0, 0].set_title('C%:' + str(np.around(CNR_tot[num]*100, decimals=2)) + ' dS:' + str(np.around(nCNR_tot[num], decimals=2)) + '\n Mean [m/s]')
            p0 = plt.colorbar(a0, ax=axarr[0, 0], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p0.ax.tick_params(labelsize=10)
            axarr[0, 0].axis('off')
            #Uncertainty
            a1 = axarr[0, 1].imshow(np.transpose(uncert['stdev'][num, ...]), cmap='hot_r', vmin = vmin_st, vmax= vmax_st)
            axarr[0, 1].set_title('Loc: ' + str(np.around(stdev_loc_tot[num], decimals=2)) + '. Total: ' + str(
                np.around(stdev_tot[num], decimals=2)) + '\nSt. dev [m/s]')
            p1 = plt.colorbar(a1, ax=axarr[0, 1], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p1.ax.tick_params(labelsize=10)
            axarr[0, 1].axis('off')
            #Bmode image
            #if MS:
            #    a2 = axarr[1, 0].imshow(b_mode_im, aspect=3.75/15, cmap='Greys_r')
            #else:
            #    a2 = axarr[1, 0].imshow(b_mode_im, aspect=3.75 / 30, cmap='Greys_r')
            #for im in axarr[1,0].get_images():
            #    im.set_clim(-70, 0)
            #axarr[1, 0].set_title('B mode image [dB]')
            #p2 = plt.colorbar(a2, ax=axarr[1, 0], fraction=0.046 * (
            #    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            #p2.ax.tick_params(labelsize=10)
            #axarr[1, 0].axis('off')
            #L*Meas
            a4 = axarr[0, 2].imshow(backp[7:, :], cmap='YlOrRd', vmin=0, vmax=1)
            # pc = axarr[0, 2].pcolormesh(np.rot90(backp[7:67, :], 2), norm=colors.PowerNorm(gamma=0.45), cmap='hot_r')
            # p4 = plt.colorbar(pc, ax=axarr[0, 2], fraction=0.046 * (67 / 64))
            axarr[0, 2].set_title('Exp: '+ data_names[num]+ '\n' + 'L*Meas', fontsize=13)
            axarr[0, 2].set_xticks([])
            axarr[0, 2].set_yticks([])
            p4 = plt.colorbar(a4, ax=axarr[0, 2], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p4.ax.tick_params(labelsize=10)
            #LBFGS
            if MS:
                a3 = axarr[1, 1].imshow(np.transpose(lbfgs_recon[:,:67,num]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            else: 
                a3 = axarr[1, 1].imshow(np.transpose(lbfgs_recon[num,...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            axarr[1, 1].set_title('C%:' + str(np.around(CNR_lbfgs[num] * 100, decimals=2)) + ' dS:' + str(
                np.around(nCNR_lbfgs[num], decimals=2)) + '\n LBFGS recon [m/s]')
            p3 = plt.colorbar(a3, ax=axarr[1, 1], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p3.ax.tick_params(labelsize=10)
            axarr[1, 1].axis('off')
            #Orig VN
            a4 = axarr[1, 0].imshow(np.transpose(vn['recon'][num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
            axarr[1, 0].set_title('C%:' + str(np.around(CNR_VN[num] * 100, decimals=2)) + ' dS:' + str(
                    np.around(nCNR_VN[num], decimals=2)) + '\nVN recon [m/s]')
            p4 = plt.colorbar(a4, ax=axarr[1, 0], fraction=0.046 * (
                    uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            p4.ax.tick_params(labelsize=10)
            axarr[1, 0].axis('off')
            #Orig mask
            a5 = axarr[1, 2].imshow((uncert['mask_out'][...,num]), cmap='gray')
            axarr[1, 2].set_title('Inc. Mask')
            axarr[1, 2].axis('off')
            axarr[1, 2].axis('off')
            axarr[1, 2].grid('off')
            #Extras
            # Measmnts corr coeff
            # meas_1 = uncert['orig']['CorrCoeff'][:, :,:].copy()
            # meas_1[np.isnan(uncert['orig']['CorrCoeff'][:, :,:])] = 0
            # meas_1 = meas_1.reshape(84, 64, 6, order='F').copy()
            # a4 = axarr[1, 2].imshow(np.mean(meas_1, 2), cmap='YlOrRd', vmin=0, vmax=1)
            # p4 = plt.colorbar(a4, ax=axarr[1, 2], fraction=0.046 * (
            #         uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            # axarr[1, 2].set_title('Mean CorrCoeff', fontsize=13)
            # axarr[1, 2].set_xticks([])
            # axarr[1, 2].set_yticks([])
            # p4.ax.tick_params(labelsize=10)
            #Masked stdev
            # a5 = axarr[1, 2].imshow((kk), cmap='hot_r', vmax=20, vmin=2)
            # axarr[1, 2].set_title(
            #     'Masked stdev: ' + str(np.around(mask_dev, decimals=2)) + '\n St. dev [m/s]',
            #     fontsize=13)
            # axarr[1, 2].set_xticks([])
            # axarr[1, 2].set_yticks([])
            # p5 = plt.colorbar(a5, ax=axarr[1, 2], fraction=0.046 * (
            #         uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            # p5.ax.tick_params(labelsize=10)
            #Measurements
            #meas = np.ones(np.shape(uncert['orig']['measmnts'][:, num]))
            #meas[np.isnan(uncert['orig']['measmnts'][:, num])] = 0
            #if MS:
            #    meas = meas.reshape(84, 64, 6, order='F').copy()
            #else:
            #    meas = meas.reshape(84, 64, 15, order='F').copy()
            # a4 = axarr[0, 2].imshow((np.sum(meas[:67,:,:], 2)), cmap='YlOrRd', vmin=0, vmax=6)
            # axarr[0, 2].set_title('Meas mask', fontsize=13)
            # axarr[0, 2].set_xticks([])
            # axarr[0, 2].set_yticks([])
            # p4 = plt.colorbar(a4, ax=axarr[0, 2], fraction=0.046 * (
            #         uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
            # p4.ax.tick_params(labelsize=10)

            #f.subplots_adjust(hspace=-0.04)
            #f.suptitle(data_names[num])
            f.subplots_adjust(wspace=0.5)
            plt.show()
    return uncert
```

#### Creating data: edditing masks, CNRs, uncertainties

```python
uncert['mask'][np.isnan(uncert['mask'])] = 0
uncert['mask_out'] = uncert['mask'].copy()
for m in range(uncert['mask'].shape[-1]):
    pos = np.where(uncert['mask'][...,m] == 1)
    for i in range(len(pos[0])):
        uncert['mask_out'][pos[0][i] - 7:pos[0][i] + 7, pos[1][i] - 7:pos[1][i] + 7,m] = 1
        #if overall == '2' or overall == 'phantom1' or overall == 'phantom2':
        #    uncert['mask_out'][pos[0][i] - 7:pos[0][i] + 7, pos[1][i] - 7:pos[1][i] + 7] = 1
        #elif overall == '3':
        #    uncert['mask'][pos[0][i] - 1:pos[0][i] + 1, pos[1][i] - 1:pos[1][i] + 1] = 1
        #    uncert['mask_out'][pos[0][i] - 8:pos[0][i] + 8, pos[1][i] - 8:pos[1][i] + 8] = 1
        #else:
        #   uncert['mask'][pos[0][i] - 2:pos[0][i] + 2, pos[1][i] - 2:pos[1][i] + 2] = 1
        #   uncert['mask_out'][pos[0][i] - 9:pos[0][i] + 9, pos[1][i] - 9:pos[1][i] + 9] = 1
        
    uncert['mask_out'][...,m] = uncert['mask_out'][...,m] - uncert['mask'][...,m]
    #Computing CNR
    pre = uncert['mean'][m,...]
    sig_in = np.mean(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.mean(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    # CNR = np.abs((sig_in - sig_out) / dev_out)
    # CNR = np.abs(sig_in - sig_out) / np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2))
    # CNR = sig_in/dev_in
    # CNR = sig_in/dev_in + sig_out/dev_out
    # Richard's implementation
    # CNR = 2 * np.power(sig_in - sig_out, 2) / (np.power(dev_in, 2) + np.power(dev_out, 2))
    CNR = np.abs(sig_in - sig_out)/sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_tot.append(CNR)
    nCNR_tot.append(nCNR)
    
    #Computing CNR LBFGS
    if MS:
        pre = lbfgs_recon[...,m]
    else:
        pre = lbfgs_recon[m, ...]
    sig_in = np.mean(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.mean(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    CNR = np.abs(sig_in - sig_out) / sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_lbfgs.append(CNR)
    nCNR_lbfgs.append(nCNR)
    
    #Computing CNR VN
    pre = vn['recon'][m, ...]
    sig_in = np.mean(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.mean(np.transpose(pre)[ np.where((uncert['mask_out'][...,m]) == 1)])
    dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m])== 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    CNR = np.abs(sig_in - sig_out) / sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_VN.append(CNR)
    nCNR_VN.append(nCNR)
    
    #Computing the uncertainty local and global
    pre = np.transpose(uncert['stdev'][m, ...])
    error_stdev = np.sum(pre)
    error_stdev_mean = np.sum(pre) / np.prod(uncert['stdev'][m, ...].shape)
    error_stdev_loc = np.sum(pre[np.where(uncert['mask'][...,m] == 1)]) / np.sum((uncert['mask'][...,m] == 1))
    #Computing normalized uncertainty, not relevant now
    #pre_2 = uncert['mean'][num, ...]
    #dev_out = np.std(np.transpose(pre_2)[np.where((uncert['mask_out']) == 1)])
    #dev_in = np.std(np.transpose(pre_2)[np.where((uncert['mask']) == 1)])
    # stdev_loc_tot.append(error_stdev_loc/np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2)))
    stdev_loc_tot.append(error_stdev_loc)
    stdev_tot.append(error_stdev_mean)
        
```

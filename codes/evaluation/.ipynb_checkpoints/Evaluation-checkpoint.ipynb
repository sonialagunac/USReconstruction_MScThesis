#### Creating data: bigger stdevs, CNR recon, CNR gt, residual, rmse

```python
for m in range(mask_sim.shape[0]):
    pos = np.where(mask_sim[m,...] == 1)
    for i in range(len(pos[0])):
        uncert['mask_out'][m, pos[0][i] - 2:pos[0][i] + 2, pos[1][i] - 2:pos[1][i] + 2] = 1
        #The lines below normalize the uncertainty, correlations didn't improve so discarded
        #pre = uncert['mean'][m, ...]
        #dev_in = np.std((pre)[np.where(uncert['mask_out'][m, ...] == 1)])
        #dev_out = np.std((pre)[np.where(uncert['mask_out'][m, ...] == 0)])
        # stdevs_big.append(np.mean(uncert['stdev'][m, uncert['mask_out'][m, ...] == 1])/ np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2)))
    
    stdevs_big.append(np.mean(uncert['stdev'][m, uncert['mask_out'][m, ...] == 1]))
    #The lines below normalize the uncertainty, correlations didn't improve so discarded
    # stdevs_big.append(np.mean(uncert['stdev'][m, uncert['mask_out'][m, ...] == 1])/np.mean(uncert['mean'][m, mask_sim[m, ...] == 1]))
    pre = uncert['mean'][m, ...]
    sig_in = np.mean((pre)[np.where((uncert['mask_out'][m,...]) == 1)])
    sig_out = np.mean((pre)[np.where(uncert['mask_out'][m,...] == 0)])
    dev_in = np.std((pre)[np.where(uncert['mask_out'][m, ...] == 1)])
    dev_out = np.std((pre)[np.where(uncert['mask_out'][m,...]  == 0)])
    # stdevs_big.append(np.mean(uncert['stdev'][m, uncert['mask_out'][m, ...] == 1]) /np.abs(sig_in - sig_out))

    # Computing contrast values
    # CNR = np.abs(sig_in - sig_out) / np.sqrt(np.power(dev_in,2) + np.power(dev_out,2))
    CNR = np.abs(sig_in - sig_out)
    # CNR = sig_in/dev_in + sig_out/dev_out
    # CNR = sig_in/dev_in
    # CNR_tot.append([CNR,np.abs(sig_in - sig_out), np.sqrt(np.power(dev_in,2) + np.power(dev_out,2))] )
    CNR_tot.append(CNR)

    # Computing the CNR of the gt data
    pre = 1 / mat['gt_slowness'][ m, ...]
    sig_in = np.mean((pre)[np.where((uncert['mask_out'][m, ...]) == 1)])
    sig_out = np.mean((pre)[np.where(uncert['mask_out'][m, ...] == 0)])
    dev_in = np.std((pre)[np.where(uncert['mask_out'][m, ...] == 1)])
    dev_out = np.std((pre)[np.where(uncert['mask_out'][m, ...] == 0)])
    # CNR = np.abs(sig_in - sig_out)/dev_out
    # CNR = np.abs(sig_in - sig_out) / np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2))
    CNR = np.abs(sig_in - sig_out)
    # CNR = sig_in/dev_in
    # CNR = sig_in/dev_in + sig_out/dev_out
    CNR_tot_gt.append(CNR)
    
uncert['CNR'] = CNR_tot.copy()
uncert['CNR_gt'] = CNR_tot_gt.copy()
for kk in range(len(stdevs)):
    if np.isnan(stdevs[kk]):
        stdevs_big[kk] = np.nan
stdevs = stdevs_big
uncert['stdevs_mean'] = stdevs_mean
uncert['stdevs'] = stdevs

uncert['residual'] = np.abs(1/mat['gt_slowness'][...] - uncert['mean'])

RMSEdrop.append(RMSE_single(mat['gt_slowness'][...], uncert['mean'], uncert['mask_out']))
uncert['rmse'] = RMSEdrop[0]
titles.append(titles_drop[0])
RMSE_tot[docs[type]] = RMSEdrop.copy()

RMSEdrop_tot.append(RMSE_single_tot(mat['gt_slowness'][...], uncert['mean']))
RMSEvn, RMSElbfgs = RMSE_duo(mat['gt_slowness'][ ...], vn['recon'], lbfgs_recon['recon_lbfgs'])
#In case you wanna compare particular VN models with the same code
# vn_path_2 = os.path.join('/scratch_net/biwidl307/sonia/USImageReconstruction-Melanie/runs/32_mix_ideal_20l_reg_mix_reg1e5/eval-vn-120000',filename[0])
# vn_2 = hdf5storage.loadmat(vn_path_2)
# RMSEvn_2, RMSElbfgs = RMSE_duo(mat['gt_slowness'][1, ...], vn_2['recon'], lbfgs_recon['recon_lbfgs'])
# RMSEdrop.append(RMSEvn_2)
# titles.append('reg1e5')
RMSEdrop_tot.append(RMSEvn)
RMSEdrop_tot.append(RMSElbfgs)
titles.append('VN')
titles.append('LBFGS')
```

### Functions

```python
def norm_data(data):
    """
    normalize data to have mean=0 and standard_deviation=1
    """
    mean_data=np.mean(data)
    std_data=np.std(data, ddof=1)
    return (data-mean_data)/(std_data)

def ncc(data0, data1):
    """
    normalized cross-correlation coefficient between two data sets

    Parameters
    ----------
    data0, data1 :  numpy arrays of same size
    """
    return (1.0/(data0.size-1)) * np.sum(norm_data(data0)*norm_data(data1))

def nll(stdev, mean, gt):
    var = np.power(stdev, 2)
    nll_i=0.5*np.log(var) + 0.5*np.power(gt-mean, 2) + 0.5*np.log(2 * np.pi)
    nll = np.mean(nll_i)
    return nll, nll_i

def RMSE_duo(img, recon, lbfgs):
    """
    Args:
        img: Ground truth img
        recon: Reconstructed VN
        lbfgs: Reconstructed LBFGS

    Returns: RMSE
    """
    RMSEvn = []
    RMSElbfgs = []
    for p in range(img.shape[0]):
        xtrue = 1 / img[p, ...]
        recon_vn = recon[p, ...]
        recon_l = lbfgs[p, ...]
        RMSEvn.append(np.sqrt(np.mean(np.power(recon_vn - xtrue, 2))))
        RMSElbfgs.append(np.sqrt(np.mean(np.power(np.transpose(recon_l) - xtrue, 2))))

    return RMSEvn, RMSElbfgs
```

```python
def RMSE_single(img, recon, mask):
    """
    Args:
        img: Ground truth img
        recon: Reconstructed

    Returns: RMSE
    """
    RMSErecon = []
    for p in range(img.shape[0]):
        # xtrue = 1 / img[p, ...]
        # recon_s = recon[p, ...]

        xtrue = 1 / img[p, mask[p, ...] == 1]
        recon_s = recon[p, mask[p, ...] == 1]
        RMSErecon.append(np.sqrt(np.mean(np.power(recon_s - xtrue, 2))) )

    return RMSErecon
def RMSE_single_tot(img, recon):
    """
    Args:
        img: Ground truth img
        recon: Reconstructed

    Returns: RMSE
    """
    RMSErecon = []
    for p in range(img.shape[0]):
        xtrue = 1 / img[p, ...]
        recon_s = recon[p, ...]
        RMSErecon.append(np.sqrt(np.mean(np.power(recon_s - xtrue, 2))) )

    return RMSErecon
```

```python
def compute_corr_small(uncert, vn, lbfgs_recon, data_names,stdev_loc_tot,stdev_tot,MS, CNR_tot,nCNR_tot,CNR_lbfgs,nCNR_lbfgs,CNR_VN,nCNR_VN,uncert_dS,metric, std_dif, std_dif_2, plot = False, samples = 3):
    uncert['mask_dev'] = []
    for nm in range(len(num_sub)):
        nm = int(nm)
        if plot:
            vmin_c = 1450
            vmax_c = 1540
            #vmin_st = 2
            #vmax_st = 20
            # vmin_c = np.amin([uncert['mean'][num, ...], vn['recon'][num, ...], np.transpose(lbfgs_recon[num,...])])
            # vmax_c = np.amax([uncert['mean'][num, ...], vn['recon'][num, ...], np.transpose(lbfgs_recon[num,...])])

            f, axarr = plt.subplots(num_sub[nm], 7, figsize = (24,18))
            for m in range(num_sub[nm]):
                num = sum(num_sub[:nm]) + m 
#                 vmin_c = np.amin([uncert['mean'][num, ...].ravel(), vn['recon'][num, ...].ravel(), np.transpose(lbfgs_recon[num,...]).ravel()])
#                 vmax_c = np.amax([uncert['mean'][num, ...].ravel(), vn['recon'][num, ...].ravel(), np.transpose(lbfgs_recon[num,...]).ravel()])
                #Computing L*Meas
                b = np.ones(np.shape(bf_im['measmnts'][:, num]))
                b[np.isnan(bf_im['measmnts'][:, num])] = 0
                #Computing L*CorrCoeff
                # b = orig_data['CorrCoeff'][:, num].copy()
                # b[np.isnan(orig_data['CorrCoeff'][:, num])] = 0
                a = np.matmul(np.transpose(np.abs(bf_im['L'])), b)
                backp = (a.reshape(84, 64, order='F').copy())
                backp = (backp - np.amin(backp[7:, :])) / (np.amax(backp[7:, :]) - np.amin(backp[7:, :]))
                if short:
                    backp = backp[:67,...]
                kk = np.transpose(uncert['stdev'][num,:, 7:]) * (1 - 0.5*backp[7:, :])
#                 mask_dev = np.mean(kk[uncert['mask'][7:,:,num] == 1])
                mask_dev = np.mean(kk[uncert['mask'][7:67,:,num] == 1])
                uncert['mask_dev'].append(mask_dev)
                #Computing Bmode intensities
                b_mode_im_pre = np.abs(hilbert(np.mean(uncert['b_mode'][..., num], 2)))
                b_mode_inter = b_mode_im_pre / np.max(np.abs(b_mode_im_pre))
                b_mode_im = 20 * np.log10(b_mode_im_pre / np.max(np.abs(b_mode_im_pre)))
                mask_res = skimage.transform.resize(uncert['mask'][...,num],np.shape(b_mode_im))
                 #Creating mask perimeter coordinates
                con = skimage.measure.find_contours(mask_res)
                con_1 = []
                con_2 = []
                for k in range(len(con[0])):
                    con_1.append(con[0][k][0])
                    con_2.append(con[0][k][1])
                #Residual computation
                recon_1 = np.transpose((bf_im_sub['L_fact']/vn['recon_long'][num,...] - vn['k'][num])/vn['std'][num])
                din_whiten = vn['din_whiten'][...,num]
                origs = np.matmul(bf_im['L'],np.ravel(recon_1,order='F'))
                origs[din_whiten == 0] = np.nan
                din_whiten[din_whiten == 0] = np.nan
                res = origs - din_whiten
                mes = np.reshape(din_whiten, [84,64,-1], order = 'F')
                p = np.reshape(origs, [84,64,-1], order = 'F')
                res = np.reshape(res, [84,64,-1], order = 'F')
                p[p==0] = np.nan
                res[np.isnan(res)] = 0
                # l = plt.imshow(res[:67,:,3],cmap = 'hot_r')
                # plt.colorbar(l)
                # plt.show()
                res = np.mean(res,axis = 2)
                #Mean recon
                a0 = axarr[m, 1].imshow(np.transpose(uncert['mean'][num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
                p0 = plt.colorbar(a0, ax=axarr[m, 1], fraction=0.046 * (uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p0.ax.tick_params(labelsize=10)
                axarr[m, 1].axis('off')
                #Uncertainty
                a1 = axarr[m, 2].imshow(np.transpose(uncert['stdev'][num, ...]), cmap='hot_r')
                                        #, vmin = vmin_st, vmax= vmax_st)
                p1 = plt.colorbar(a1, ax=axarr[m, 2], fraction=0.046 * (uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p1.ax.tick_params(labelsize=10)
                axarr[m, 2].axis('off')
                #L*Meas
                a4 = axarr[m, 3].imshow(backp[7:, :], cmap='YlOrRd', vmin=0, vmax=1)
                # pc = axarr[0, 2].pcolormesh(np.rot90(backp[7:67, :], 2), norm=colors.PowerNorm(gamma=0.45), cmap='hot_r')
                # p4 = plt.colorbar(pc, ax=axarr[0, 2], fraction=0.046 * (67 / 64))
                axarr[m, 3].set_xticks([])
                axarr[m, 3].set_yticks([])
                p4 = plt.colorbar(a4, ax=axarr[m, 3], fraction=0.046 * (
                        uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p4.ax.tick_params(labelsize=10)
                #LBFGS
                if MS:
                    a3 = axarr[m, 6].imshow(np.transpose(lbfgs_recon[:,:67,num]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
                else: 
                    a3 = axarr[m, 6].imshow(np.transpose(lbfgs_recon[num,...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
                
                p3 = plt.colorbar(a3, ax=axarr[m, 6], fraction=0.046 * (
                        uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p3.ax.tick_params(labelsize=10)
                axarr[m, 6].axis('off')
                #Orig VN
                a6 = axarr[m, 5].imshow(np.transpose(vn['recon'][num, ...]), cmap='turbo', vmin=vmin_c, vmax=vmax_c)
                p6 = plt.colorbar(a6, ax=axarr[m, 5], fraction=0.046 * (
                        uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p6.ax.tick_params(labelsize=10)
                axarr[m, 5].axis('off')
                #Bmode image
                if MS:
                   a6 = axarr[m, 0].imshow(b_mode_im, aspect=3.75/15, cmap='Greys_r')
                else:
                   a6 = axarr[m, 0].imshow(b_mode_im, aspect=3.75 / 30, cmap='Greys_r')
                for im in axarr[m,0].get_images():
                   im.set_clim(-70, 0)
                #axarr[1, 0].set_title('B mode image [dB]')
                axarr[m, 0].plot(con_2,con_1, '--', linewidth=2, color='firebrick')
                p6 = plt.colorbar(a6, ax=axarr[m, 0], fraction=0.046 * (
                   uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p6.ax.tick_params(labelsize=10)
                axarr[m, 0].axis('off')
                #Residual
                a4 = axarr[m, 4].imshow(res[:67,:],cmap = 'hot_r')
                p4 = plt.colorbar(a4, ax=axarr[m, 4], fraction=0.046 * (uncert['mean'][num, ...].shape[1] / uncert['mean'][num, ...].shape[0]))
                p4.ax.tick_params(labelsize=10)
                axarr[m, 4].axis('off')
                #Orig mask
#                 a5 = axarr[m, 2].imshow((uncert['mask_out'][...,num]), cmap='gray')
#                 axarr[m, 2].axis('off')
#                 axarr[m, 2].axis('off')
                if nm == 0 and m ==0:
                    axarr[m, 0].set_title('Exp: '+ data_names[num]+ '\n B mode image [dB]')
#                     axarr[m, 2].set_title('Exp: '+ data_names[num]+ '\n Inc. Mask')
                    axarr[m, 3].set_title('L*Meas')
                    axarr[m, 4].set_title('Residual')
                    axarr[m, 2].set_title('Loc. std: ' + str(np.around(stdev_loc_tot[num], decimals=2)) + '. Δuncert: ' + str(
                    np.around(uncert_dS[num], decimals=2)) 
#                                           + '\nStd, st_dif:' +str(np.around(std_dif[num], decimals=2))
#                                            + ' st_dif2:' +str(np.around(std_dif_2[num], decimals=2)) 
                                         )
#                                          
                    axarr[m, 1].set_title(
#                         'C%:' + str(np.around(CNR_tot[num]*100, decimals=2)) + 
                          ' Mean [m/s]' + '\n  ΔSoS: ' + str(np.around(nCNR_tot[num], decimals=2)) 
#                         + ', mSoS:' + str(np.around(metric[num], decimals=2))
                    )
                    axarr[m, 6].set_title(
#                         'C%:' + str(np.around(CNR_lbfgs[num] * 100, decimals=2)) + 
                         'LBFGS recon [m/s]' + '\n ΔSoS: ' + str(np.around(nCNR_lbfgs[num], decimals=2)))
                    axarr[m, 5].set_title(
#                         'C%:' + str(np.around(CNR_VN[num] * 100, decimals=2)) +
                        'VN recon [m/s]' +  '\nΔSoS: ' + str(np.around(nCNR_VN[num], decimals=2)) )
        
                else:
                    axarr[m, 0].set_title('Exp: '+ data_names[num])
                    axarr[m, 2].set_title('Loc. std: ' + str(np.around(stdev_loc_tot[num], decimals=2)) + '. Δuncert: ' + str(
                    np.around(uncert_dS[num], decimals=2)) 
#                                           + 'stdif:' +str(np.around(std_dif[num], decimals=2))
#                                           + ' st_dif2:' +str(np.around(std_dif_2[num], decimals=2)))
                                         )
                    axarr[m, 1].set_title(
#                         'C%:' + str(np.around(CNR_tot[num]*100, decimals=2)) + 
                        ' ΔSoS: ' + str(np.around(nCNR_tot[num], decimals=2)) + ' [m/S]'
#                         + 'mSoS:' + str(np.around(metric[num], decimals=2))
                    )
                    axarr[m, 6].set_title(
#                         'C%:' + str(np.around(CNR_lbfgs[num] * 100, decimals=2)) + 
                        ' ΔSoS: ' + str(
                    np.around(nCNR_lbfgs[num], decimals=2))+ ' [m/S]')
                    axarr[m, 5].set_title(
#                         'C%:' + str(np.around(CNR_VN[num] * 100, decimals=2)) + 
                        ' ΔSoS: ' + str(
                        np.around(nCNR_VN[num], decimals=2))+ ' [m/S]')
           # f.subplots_adjust(wspace=0.5)
            pk = 0
            while os.path.isfile('/scratch_net/biwidl307/sonia/kk/clinical/dilated_clinical_bay_CA_VS' + str(nm) + str(MS) + str(pk) + '.pdf'):
                pk += 1
            plt.savefig('/scratch_net/biwidl307/sonia/kk/clinical/dilated_clinical_bay_CA_VS' + str(nm) + str(MS) + str(pk) + '.pdf')
            plt.show()
    return uncert
```

#### Creating data: edditing masks, CNRs, uncertainties

```python
CNR_tot = []
nCNR_tot = []
CNR_VN = []
nCNR_VN = []
CNR_lbfgs = []
nCNR_lbfgs = []
std_dif = []
std_dif_2 = []
stdev_tot = []
stdev_loc_tot = []
noise_tot = []
uncert['mask'][np.isnan(uncert['mask'])] = 0
uncert['mask_out'] = uncert['mask'].copy()
kernel = skimage.morphology.disk(8)
for m in range(uncert['mask'].shape[-1]):
#     pos = np.where(uncert['mask'][...,m] == 1)
#     for i in range(len(pos[0])):
# #         uncert['mask_out'][pos[0][i] - 7:pos[0][i] + 7, pos[1][i] - 7:pos[1][i] + 7,m] = 1
#         uncert['mask_out'][pos[0][i] - 8:pos[0][i] + 8, pos[1][i] - 8:pos[1][i] + 8,m] = 1
#         #if overall == '2' or overall == 'phantom1' or overall == 'phantom2':
#         #    uncert['mask_out'][pos[0][i] - 7:pos[0][i] + 7, pos[1][i] - 7:pos[1][i] + 7] = 1
#         #elif overall == '3':
#         #    uncert['mask'][pos[0][i] - 1:pos[0][i] + 1, pos[1][i] - 1:pos[1][i] + 1] = 1
#         #    uncert['mask_out'][pos[0][i] - 8:pos[0][i] + 8, pos[1][i] - 8:pos[1][i] + 8] = 1
#         #else:
#         #   uncert['mask'][pos[0][i] - 2:pos[0][i] + 2, pos[1][i] - 2:pos[1][i] + 2] = 1
#         #   uncert['mask_out'][pos[0][i] - 9:pos[0][i] + 9, pos[1][i] - 9:pos[1][i] + 9] = 1
        
#     uncert['mask_out'][...,m] = uncert['mask_out'][...,m] - uncert['mask'][...,m]
    img_dilation = binary_dilation(uncert['mask'][...,m], kernel) 
    uncert['mask_out'][...,m] = img_dilation - uncert['mask'][...,m]
    #Computing CNR
    pre = uncert['mean'][m,...]
    pre_l = uncert['mean_long'][m,...]
    sig_in = np.median(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.median(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
#     sig_out = np.mean(pre_l[:,-1])
    dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    # CNR = np.abs((sig_in - sig_out) / dev_out)
    # CNR = np.abs(sig_in - sig_out) / np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2))
    noise = sig_in/dev_in
    # CNR = sig_in/dev_in + sig_out/dev_out
    # Richard's implementation
    # CNR = 2 * np.power(sig_in - sig_out, 2) / (np.power(dev_in, 2) + np.power(dev_out, 2))
    CNR = np.abs(sig_in - sig_out)/sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_tot.append(CNR)
    nCNR_tot.append(nCNR)
    noise_tot.append(noise)
    
    #Computing CNR LBFGS
    if MS:
        pre = lbfgs_recon[...,m]
        pre_l = lbfgs_recon_long[...,m]
    else:
        pre = lbfgs_recon[m, ...]
        pre_l = lbfgs_recon_long[m,...]
    sig_in = np.median(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
#     sig_out = np.mean(pre_l[:,-1])
    sig_out = np.median(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
#     dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m]) == 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    CNR = np.abs(sig_in - sig_out) / sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_lbfgs.append(CNR)
    nCNR_lbfgs.append(nCNR)
    
    #Computing CNR VN
    pre = vn['recon'][m, ...]
    pre_l = vn['recon_long'][m,...]
    sig_in = np.median(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.median(np.transpose(pre)[ np.where((uncert['mask_out'][...,m]) == 1)])
#     sig_out = np.mean(pre_l[:,-1])
#     dev_out = np.std(np.transpose(pre)[np.where((uncert['mask_out'][...,m])== 1)])
    dev_in = np.std(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    #Definition of CNR alternates
    CNR = np.abs(sig_in - sig_out) / sig_out
    nCNR = np.abs(sig_in - sig_out)
    CNR_VN.append(CNR)
    nCNR_VN.append(nCNR)
    
    #Computing the uncertainty local and global
    pre = np.transpose(uncert['stdev'][m, ...])
    error_stdev = np.sum(pre)
    error_stdev_mean = np.sum(pre) / np.prod(uncert['stdev'][m, ...].shape)
    error_stdev_loc = np.sum(pre[np.where(uncert['mask'][...,m] == 1)]) / np.sum((uncert['mask'][...,m] == 1))
    #Computing normalized uncertainty, not relevant now
    #pre_2 = uncert['mean'][num, ...]
    #dev_out = np.std(np.transpose(pre_2)[np.where((uncert['mask_out']) == 1)])
    #dev_in = np.std(np.transpose(pre_2)[np.where((uncert['mask']) == 1)])
    # stdev_loc_tot.append(error_stdev_loc/np.sqrt(np.power(dev_in, 2) + np.power(dev_out, 2)))
    stdev_loc_tot.append(error_stdev_loc)
    stdev_tot.append(error_stdev_mean)
    #Computing uncertainty diference
    pre = uncert['stdev'][m, ...]
    pre_l = uncert['stdev_long'][m,...]
    sig_in = np.mean(np.transpose(pre)[np.where((uncert['mask'][...,m]) == 1)])
    sig_out = np.mean(np.transpose(pre)[ np.where((uncert['mask_out'][...,m]) == 1)])
    sig_out_2 = np.mean(pre_l[:,-1])
    std_dif.append(np.abs(sig_out - sig_in))
    std_dif_2.append(np.abs(sig_out_2 - sig_in))
        
#Intermediate for later compute corr
uncert_dS = stdev_loc_tot
metric = nCNR_tot
```
